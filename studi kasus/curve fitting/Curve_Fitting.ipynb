Misal kita mempunyai suatu persamaan yang ingin dicocokkan kedalam suatu set data. Apakah parameter terbaik yang dapat digunakan dalam persamaan tersebut sehingga persamaan tersebut dapat merepresentasikan set data secara optimum?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#data kedalaman dan temperatur
x = np.array([60,80,100,120,140,160,180,200])
t = np.array([0.5147,0.5151,0.5155,0.5161,0.5167,0.5175,0.5183,0.5192])

#visualisasi
plt.scatter(x,t,label='data')
plt.xlabel('Kedalaman')
plt.ylabel('Temperatur')
plt.legend()
plt.show()
```

Jika terdapat dua persamaan yaitu:
1.   t = ax + b
2.   t<sup>2</sup> = ax<sup>2</sup> + b

tentukan persamaan dan parameter dengan misfit paling minimum!

```python
#persamaan 1
def f(x, a, b):
   return a*x + b

#persamaan 2
def g(x, a, b):
   return a*(x**2) + b
```

```python
#curve fitting persamaan 1
popt,cov = curve_fit(f, x, t)
a,b = popt

#membuat set baru berdasarkan hasil curve fitting
x_baru = np.arange(min(x), max(x))
t_baru = f(x_baru, a, b)

#rmse
dt=f(x, a, b)-t
dt_=dt**2
e=sum(dt)
rmse=np.sqrt(e)

#visualisasi
plt.scatter(x,t,label='data')
plt.plot(x_baru,t_baru,'r',label='curve')
plt.title('Persamaan 1 : a='+str(round(a,2))+',b='+str(round(b,2))+',RMSE ='+str(rmse))
plt.xlabel('Kedalaman')
plt.ylabel('Temperatur')
plt.legend()
plt.show()
```

```python
#curve fitting persamaan 2
popt,cov = curve_fit(g, x, t)
a,b = popt

#membuat set baru berdasarkan hasil curve fitting
x_baru = np.arange(min(x), max(x))
t_baru = g(x_baru, a, b)

#rmse
dt=g(x, a, b)-t
dt_=dt**2
e=sum(dt)
rmse=np.sqrt(e)

#visualisasi
plt.scatter(x,t,label='data')
plt.plot(x_baru,t_baru,'r',label='curve')
plt.title('Persamaan 2 : a='+str(round(a,2))+',b='+str(round(b,2))+',RMSE ='+str(rmse))
plt.xlabel('Kedalaman')
plt.ylabel('Temperatur')
plt.legend()
plt.show()
```

terlihat bahwa persamaan kedua memiliki error terkecilm sehingga dapat disimpulkan bahwa data tersebut lebih cocok dengan persamaan dua, dengan nilai a=0, b=0.51
