```python
import pandas as pd
#Pandas merupakan modul yang sering digunakan untuk membaca data, karena mudah dan metode-metodenya yang sederhana
import numpy as np
import matplotlib.pyplot as plt
```

Akan membaca dua katalog gempa dan menggabungkannya menjadi satu. Pembacaan perlu dilakukan perubahan header sehingga penggabungan bisa dilakukan

```python
data_USGS = pd.read_csv('Data USGS.txt',sep="\t")
data_USGS = data_USGS.rename(columns={'latitude': 'Lat', 'longitude': 'Lon', 'mag':'Mag','depth':'Depth','Seconds':'Second'})

data_EMSC = pd.read_csv('Data EMSC.txt',sep="\t")
data_EMSC = data_EMSC.rename(columns={'Latitude': 'Lat', 'Longitude': 'Lon', 'Magnitude':'Mag','Mont':'Month'})
```

```python
#untuk menggabungkan kedua data
df=[data_USGS,data_EMSC]
df = pd.concat(df)

#melihat data
df
```

```python
#mengurutkan data berdasarkan tahun, bulan, hari, jam, menit dan detik
df['M'] = df['M'].astype(str)
df = df.sort_values(by=['Year', 'Month', 'Day', 'Hour' , 'Minute','Second'])
df = df.reset_index(drop=True)

df
```

Terlihar di kolom M, terdapat perbedaan penulisan. Terdapat data yang tertulis dengan M besar ataupun m kecil, B besar atau b kecil, dst. Ini perlu diperbaiki sehingga pengolahan lebih mudah dilakukan.

```python
#mengubah data menjadi bentuk list, sehingga lebih mudah untuk diolah
datalist = df.values.tolist()
kolom = df.columns
index = range(0,len(kolom))
header = dict(zip(kolom,index))

#algoritma perubahan data
for i in range(len(datalist)):
  if 'm' in datalist[i][header['M']].lower():
    if 'l' in datalist[i][header['M']].lower():
      datalist[i][header['M']] = 'Ml'
    elif 's' in datalist[i][header['M']].lower():
      datalist[i][header['M']] = 'Ms'
    elif 'w' in datalist[i][header['M']].lower():
      datalist[i][header['M']] = 'Mw'
    elif 'b' in datalist[i][header['M']].lower():
      datalist[i][header['M']] = 'Mb'

#mengembalikann list menjadi dataframe pandas
database = pd.DataFrame(datalist, columns = df.columns)
database = database.dropna()
database
```

Jika diketahui bahwa urutan kepercayaan data dari yang terpercaya adalah EMSC kemudian USGS maka dapat tambahkan urutan berdasarkan 

```python
#mengubah data menjadi angka, sehingga lebih mudah untuk diurut
database['Type'] = database['Type'].replace(['USGS','EMSC',],[1,2])
database = database.sort_values(by=['Year', 'Month', 'Day', 'Hour' , 'Minute','Second','Type'])
database = database.reset_index(drop=True)

#mengembalikan nama katalog menjadi nama
database['Type'] = database['Type'].replace([1,2],['USGS','EMSC'])

database
```
